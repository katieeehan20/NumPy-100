```python
# speed comparison b/w np.array and list by using timer

# a,b = array
# A,B = list

from timeit import default_timer as timer

a=np.random.randn(1000)
b=np.random.randn(1000)

A=list (np.random.randn(1000))
B=list (np.random.randn(1000))
T=1000
```

```python
a
```

```python
A
```

```python
# function to use for numpy

def dot1():
    return a.dot(b)
```

```python
# function to use for list

def dot2():
    dot=0
    for i in range(len(a)):
        dot += a[i]*b[i]
    return dot
```

```python
start=timer()
for t in range(T):
    dot1()
end=timer()
t1=end-start
```

```python
start=timer()
for t in range(T):
    dot2()
end=timer()
t2=end-start
```

```python
print(t1)
```

```python
print(t2)
```

```python
# you can see how much faster array is than list

print(t2/t1)
```

```python
# solving linear systems

# x1+x2=2200
# 1.5x1 + 4x2 = 5050

a= np.array([[1,1], [1.5, 4]])
b= np.array([2200, 5050])

x= np.linalg.solve(a,b)
x
```
